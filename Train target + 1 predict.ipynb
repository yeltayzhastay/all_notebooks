```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense

import tensorflow as tf
import matplotlib as mpl
import matplotlib.pyplot as plt
import os

mpl.rcParams['figure.figsize'] = (8, 6)
mpl.rcParams['axes.grid'] = False
```

```python
#%%
df = pd.read_csv("C:/Users/Zhastay/Downloads/jena_climate_2009_2016/temp10x.csv")
#%%
df['t8'][9] = 25
df['t8'][14] = 25
```

```python
df = df.set_index('id')
```

```python
df.columns = ['t1', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9', 't10']
```

```python
df['t1+1'] = df['t1'].shift(-1)
```

```python
df['t1+1'].fillna(24, inplace=True)
```

```python
df
```

```python
#%%
x = df.drop('t1+1', axis = 1)
y = df['t1+1']
#%%
y=np.reshape(y.values, (-1,1))

#%%
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
print(scaler_x.fit(x))
xscale=scaler_x.transform(x)
print(scaler_y.fit(y))
yscale=scaler_y.transform(y)
#%%
X_train, X_test, y_train, y_test = train_test_split(xscale, yscale)
```

```python
model = Sequential()
model.add(Dense(16, input_dim=10, kernel_initializer='normal', activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
# Use a custom metricfunction

model.compile(loss='mse', optimizer='adam', metrics=['mse'])

fit = model.fit(X_train, y_train, epochs=200,  verbose=1, validation_split=0.1)
```

```python
print(X_test)
ynew= model.predict(X_test)
ynew = scaler_y.inverse_transform(ynew)
print(" Predicted=%s" % (ynew))
my_test = scaler_y.inverse_transform(y_test)
```

```python
np.append(my_test, ynew, axis=1)
```

```python
def plot_train_history(history, title):
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  epochs = range(len(loss))

  plt.figure()

  plt.plot(epochs, loss, 'b', label='Training loss')
  plt.plot(epochs, val_loss, 'r', label='Validation loss')
  plt.title(title)
  plt.legend()

  plt.show()
```

```python
plot_train_history(fit,
                   'Visualization losses')
```
