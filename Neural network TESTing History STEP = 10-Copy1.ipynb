```python
def plot_train_history(history, ran, title):
  loss = history

  epochs = []
  for i in ran:
    epochs.append(i)

  plt.figure()

  plt.plot(epochs, loss, 'b', label="Brg1x_Vib")
  plt.title(title)
  plt.legend()

  plt.show()
```

```python
mpl.rcParams['figure.figsize'] = (100, 6)
mpl.rcParams['axes.grid'] = False
```
